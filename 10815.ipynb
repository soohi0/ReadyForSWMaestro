```python
#이진탐색 구현해야한다.
def TorF(num, length, arr):
    start = 0
    end_ = length-1
    while end_ - start >= 0:
        mid = (start + end_) // 2
        if num == arr[mid]:
            return 1
        elif num > arr[mid]:
            start = mid+1
        else:
            end_ = mid-1
        
    return 0
length = int(input())
arr = [int(i) for i in input().split(' ')]
arr.sort()
checklen = int(input())
checkarr = [int(j) for j in input().split(' ')]
check = [0 for _ in range(500000)]
c = 0
for k in checkarr:
    check[c] = TorF(k, length, arr)
    c += 1
string = ""
for l in range(checklen):
    string += str(check[l])
    string += " "
print(string) 
```
